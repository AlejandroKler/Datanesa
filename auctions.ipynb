```python
import pandas as pd
import numpy as np
import matplotlib as plt

%matplotlib inline
```

```python
df = pd.read_csv("auctions.csv", dtype={"platform":"int8", "ref_type_id":"int8", "source_id":"int8"})
```

```python
df.head()
```

```python
platform_id = pd.DataFrame(df["platform"])
platform_id["cant"] = 1
platform_cant = platform_id.groupby("platform").agg({"cant":"count"})
```

```python
platform_cant
```

```python
df_time = pd.DataFrame(df["date"])
df_time["date"] = pd.to_datetime(df_time["date"])
```

```python
df_time.date.isnull().value_counts()
```

```python
df_time["cantidad"] = 1
df_time["fecha"] = df_time["date"].dt.date
```

```python
df_time.groupby("fecha").agg({"cantidad":"sum"}).plot(figsize=(20,10));
```
