```python
import pandas as pd
import numpy as np
import matplotlib as plt

%matplotlib inline
```

```python
clicks_dtypes = {'advertiser': np.int8, 'action_id': np.float16, 'source_id': np.int8, 'country_code': 'category',
                 'latitude': np.float16, 'longitude': np.float16, 'carrier_id': np.float16, 'brand': np.float16,
                 'touchX': np.float16, 'touchY': np.float16, 'ref_hash': "int64"}

clicks = pd.read_csv('clicks.csv', dtype = clicks_dtypes)

clicks.drop('action_id', axis = 1, inplace = True) #todos NaN
clicks.drop('wifi_connection', axis = 1, inplace = True) #todos False

clicks["created"] =  pd.to_datetime(clicks["created"], errors = "coerce") #parseo de fechas
```

```python
installs_dtypes = {'application_id': 'category', 'ref_type': 'category', 'device_countrycode': 'category',
                   'device_brand': 'category', 'device_model' : 'category', 'session_user_agent': 'category',
                   'user_agent': 'category', 'kind': 'category', 'device_language': 'category'}

installs = pd.read_csv('installs.csv', dtype = installs_dtypes)
installs_mostrar = pd.read_csv("installs.csv", dtype = installs_dtypes)

installs["created"] =  pd.to_datetime(installs["created"], errors = "coerce") #parseo de fechas

installs.drop('attributed', axis = 1, inplace = True) #son todas false
installs.drop('click_hash', axis = 1, inplace = True) #son todas NaN
```

```python
auctions_dtypes = {'auction_type_id': np.float16, 'country': 'category', 'platform': 'category',
                   'ref_type_id': 'category', 'source_id': 'category'}

auctions = pd.read_csv('auctions.csv', dtype = auctions_dtypes)

auctions.drop('auction_type_id', axis = 1, inplace = True)  #todas Nan

auctions["date"] =  pd.to_datetime(auctions["date"], errors = "coerce")
```

```python
auctions["cantidad_subastas"] = 1
installs["cantidad_instalaciones"] = 1
clicks["cantidad_clicks"] = 1
```

```python
group_auctions = auctions.groupby("device_id").agg({"cantidad_subastas": "sum"}).reset_index()
group_installs = installs.groupby("ref_hash").agg({"cantidad_instalaciones": "sum"}).reset_index()
group_clicks = clicks.groupby("ref_hash").agg({"cantidad_clicks": "sum"}).reset_index()
```

```python
merge_auctions_clicks = group_auctions.merge(group_clicks, left_on="device_id", right_on="ref_hash")
```

```python
merge_auctions_clicks = merge_auctions_clicks[["device_id", "cantidad_subastas", "cantidad_clicks"]].sort_values("cantidad_clicks", ascending=False)
merge_auctions_clicks
```

```python
merge_final = merge_auctions_clicks.merge(group_installs, on="ref_hash")
```

# A qué hora se suele hacer más clicks?

```python
clicks["hour"] = clicks["created"].dt.hour
```

```python
group_clicks = clicks.groupby("hour").agg({"cantidad_clicks":"sum"})
```

# Analizo installs

```python
group_installs = installs.groupby("hour").agg({"cantidad_instalaciones":"sum"})
```

```python
group_clicks
group_installs.reset_index(inplace=True)
group_clicks["cantidad_instalaciones"] = group_installs["cantidad_instalaciones"]
```

```python
import matplotlib.pyplot as plt
```

```python
group_clicks["relacion"] = group_clicks["cantidad_instalaciones"]/group_clicks["cantidad_clicks"]
group_clicks.head()
```

```python
ax = group_clicks.plot(kind="line",figsize=(15,10),fontsize=15 , title="Cantidad de instalaciones sobre clicks", marker="o")
ax.set_xlabel("Hora del día")
ax.set_ylabel("Cantidad de instalaciones y clicks")
ax;
```

```python
relacion_clicks_installs = pd.DataFrame(group_clicks["relacion"])
```

```python
ax = relacion_clicks_installs.plot(kind="bar",figsize=(15,10),fontsize=15 , title="Relación de la cantidad de instalaciones sobre cantidad de clicks")
ax.set_xlabel("Hora del día")
ax.set_ylabel("Relación")
ax;
```
